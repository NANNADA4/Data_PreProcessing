```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("C:/MyWorkspace/Python_MachineLearning/Data_PreProcessing/titanic.csv")
titanic_df.head()
```

```python
titanic_df.info()
```

```python
titanic_df.drop(columns=['Cabin', 'Ticket', 'Embarked', 'PassengerId'], inplace=True)
titanic_df.head(5)
```

```python
titanic_df.tail(10)
```

```python
titanic_df[titanic_df.duplicated()]
```

```python
titanic_df.drop_duplicates(inplace=True)
titanic_df[titanic_df.duplicated()]
```

```python
titanic_df.tail(10)
```

```python
titanic_df.isnull().head(10)
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df.isnull().sum()
```

```python
titanic_df.rename(columns={'SibSp': 'Sinling.spouse'}, inplace=True)
titanic_df.rename(columns={'PassengerId': 'passengerid', 'Survived': 'survived', 'Pclass': 'pclass', 'Name': 'name', 'Gender': 'gender', 'Age': 'age', 'Parch': 'parch', 'Fare': 'fare'}, inplace=True)
```

```python
titanic_df.replace({'gender': {'male': 0, 'female': 1}}, inplace=True)
titanic_df.head(10)
```

```python
titanic_df.describe()
```

```python
 from sklearn.preprocessing import MinMaxScaler
 scaler = MinMaxScaler()
 titanic_df[['fare', 'gender']] = scaler.fit_transform(titanic_df[['fare', 'gender']])
```

```python
titanic_df.describe()
```

```python
titanic_df['fare'].plot(kind = "box")
```

```python
titanic_df.drop(titanic_df[titanic_df.fare >= 0.5].index, inplace=True)
titanic_df['fare'].plot(kind = "box")
```

```python
titanic_df.corr(method='pearson')
```
